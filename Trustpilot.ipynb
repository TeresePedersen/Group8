```python
import scraping_class
logfile = 'log.csv'## name your log file.
connector = scraping_class.Connector(logfile)

import requests,os,time
def ratelimit():
    "A function that handles the rate of your calls."
    time.sleep(0.33) # sleep one second.

from tqdm import tqdm_notebook as tqdm # Module for tracking progress. conda install -c conda-forge tqdm
import json
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import json

lst = ['saxo.com', 'www.elgiganten.dk', 'www.nemlig.com', 'www.bilka.dk', 'shopping.coop.dk', 'www.dsb.dk', 'www.wupti.com', 'www.boozt.com', 'sweetdeal.dk', 'www.just-eat.dk']
samlet = []
domainUrl = 'https://dk.trustpilot.com/review/'
#Loop der l칮ber alle hjemmesiderne igennem
for site in lst:
    siteUrl = domainUrl + site
    for i in range(1): #Hvor meget data skal vi bruge??
        response = requests.get(siteUrl + "?page=" + str(i)) #loop over sidenummer, page= i
        if response.ok: #HTTP kode = 200, hvis siden eksisterer
            html = response.text
            
            soup = BeautifulSoup(html,"lxml") #Henter frar response og finder article (tabeller), lxml fjerner den r칮de advarsel
            reviews = soup.find_all("article", class_="review") #Finder omr친det hvor data er, "article", med underc췂klassen "review"
        
            for rev in reviews:
                div = rev.find("div", class_="star-rating star-rating--medium") #finder sjterne kommentar
                star = div.find("img") #fjerner overfl칮delig tekst
            
                starText = star.get("alt") #Fjerner mere teskt, s친 kuin stjerner er tilbage
                fullPersonReview = site.upper() + "\n" + rev.find(class_="consumer-information__name").text.strip() + "\n" + rev.find(class_="review-content__text").text.strip() + "\n" + starText
                print(fullPersonReview)
                time.sleep(0.33)
                samlet.append(fullPersonReview)
                #fullPersonReview.append(json.loads(url.read().decode())
        else:
            print('error: site not available')

```

```python
df = pd.DataFrame(samlet)
type(df)
df.columns = ['Website']
```
